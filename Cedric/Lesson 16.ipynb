```python
import random

symbols = ["A", "B", "P", "*"]


value = 0
values = []


print("Please deposit money you would like to play with: ")
playermoney = int(input())
print("Enter stake amount: ")
stamount = int(input())
print()
for i in range(4):
    s1 = random.choices(symbols, weights=(45, 35, 15, 5))
    s2 = random.choices(symbols, weights=(45, 35, 15, 5))
    s3 = random.choices(symbols, weights=(45, 35, 15, 5))
    if s1 == s2 and s2 == s3:
        if s1 == "A":
            value = value + 0.4
        elif s1 == "B":
            value = value + 0.6
        elif s1 == "P":
            value = value + 0.8
    values.append(value)
    print(s1, s2, s3)
        
for i in range(len(values)):
    value = value + values[i-1]
value = value*stamount

print("You have won: ", value)
print("Current balance is: ", playermoney-stamount+value)
print(values)
```
