```python
import random

symbols = ["A", "B", "P", "*"]
win_amount = 0
value = 0
win_combinations = ['AAA','BBB','PPP','A*A','*AA','AA*','*A*','B*B','*BB','BB*','*B*','P*P','*PP','PP*','*P*']

print("Please deposit money you would like to play with: ")
playermoney = int(input())

while playermoney > 0:
    print("Enter stake amount: ")
    stamount = int(input())
    print()
    for i in range(4):
        s1 = random.choices(symbols, weights=(45, 35, 15, 5), k=3)
        a = ''.join(s1)
        print(a)
        if a in win_combinations:
            for j in a:
                if j == "A":
                    value = value + 0.4
                elif j == "B":
                    value = value + 0.6
                elif j == "P":
                    value = value + 0.8
                elif j =='*':
                    value = value + 0
                    
    win_amount = value * stamount
    print("You have won:", round(win_amount))
    print("Current balance is:", round(playermoney-stamount+win_amount))
    playermoney = round(playermoney-stamount+win_amount)
    win_amount = 0
    value = 0
else:
    print("You're out of balance")
```
