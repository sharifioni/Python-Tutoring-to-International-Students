```python
#Step 4
import random
word_list = ["aardvark", "baboon", "camel"]
chosen_word = random.choice(word_list)
print(chosen_word)
display = []
for i in range(len(chosen_word)):
    display.append("_")
lives = 6
    
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
    
while True:
    if "_" in display:
        guess = input("Guess a random letter from the alphabet: ").lower()
        for i in range(len(display)):
            if guess == chosen_word[i-1]:
                display[i-1] = guess
        if guess in chosen_word:
            print(f"{guess} was in the word")
            print(stages[lives])
        else:
            lives -= 1
            print(stages[lives])
            print(f"{guess} wasn't in the word")
        print(" ".join(display))
        if lives == 0:
            print("You lose")
            break
        print()
    else:
        print("Congratulations, You've guessed the word!")
        break
```
