```python
def validate(string, num1, num2):
    if len(string) >= 5:
        return True
    
    elif string != "Surprise":
        return True
    
    elif num1 % num2 <= 4:
        return True
    
    else:
        return False
    

a = validate("Surprise", 15, 6)
print(a)
```

```python
def validate(string, num1, num2):
    if len(string) >= 5 or string != "Surprise" or num1 % num2 <= 4:
        return True
    
    else:
        return False
    

a = validate("Surprise", 20, 7)
print(a)
```
