```python
#for loop
#for loop runs on iterables and sequences (strings, lists,tuples,dictionaries,sets,range)
#Loop and will run until the sequence is finished

for a in "Programming":    #char='p'
    print(a,end="-")
```

```python
grades=[67,90,85,75]
sum1=0
avg=0
n=0
for values in grades:
    sum1+=values    #Augmented Assignment                      sum1=sum1+values
    n+=1            #n=n+1
    
print(sum1)
print(n)
avg=sum1/n

print("The average of all of the course marks is",avg)
    
    
```

```python
grades=[67,90,85,75]
sum1=sum(grades)
n=len(grades)
avg=sum1/n
print("The average of all of the course marks is",avg)
```

```python
#range function
#range(start value,stop value,step) positive and negative integers
for i in range(10):    #5  , 14
    print(i)
    
```

```python
#Table of 12
#12   *     1    =     12
#12   *     2    =     24
for i in range(1,11):            
    print(f"{12:<5} *{i:>5} = {12*i:^5}") # < left alighnment   > right alighnment    ^ centre alighnment
```

```python
#Nested for loops
#for loop inside another for loop

for i in range(10):     #0 to 9     i=0
    
    for j in range(11,20):  # 11 to 19    j=11  , j=12  , j=13 , j=20
        print(i,j)
    
print("-"*50)
        
```

```python
#Lets suppose we want tables between 20 to 35   
#Nested for loops    #150 times
for i in range(20,36):    #15 values
    for j in range(1,11):  #10 values
        print(f"{i:>5} {'*':>5} {j:>5} {'=':>5} {i*j:>5}")
    print("-"*50)
        
```
