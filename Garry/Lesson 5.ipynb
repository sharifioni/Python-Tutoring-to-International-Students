Print a board 9 X 9 with numbers between 1 and 9 generated randomly
Similar to sudoku game but without applying the rules of the Sudoku 
game. For example, you can have the number 1 appear more than once 
in the row, column or diagonal

```python
import random
board=[]

for i in range(1,10):
    a=[]
    for j in range(1,10):
        a.append(random.randint(1,9))
    board.append(a)
    

for i in board:
    for j in i:
        print(j,end="\t")
    print()

    

```
